  # 111) Weather Temperature Trend (Gorakhpur)

*   Plot the temperature trend over a week in Gorakhpur using a line graph

*   This helps visualize how the temperature changes day by day.
*   Topics
     *  Line plot

    *  Labels (xlabel, ylabel)

   *   Title

    *   Grid

    *  Color and Marker style



```python
import matplotlib.pyplot as plt

#Data
days =['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature= [32, 34, 35, 33, 31, 30, 29]


plt.plot(days, temperature, color='red', marker='o' , linestyle='-', linewidth=2)


plt.title("Weekly Temperature Trend - Gorakhpur")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")

plt.grid(True)

plt.show()
```

 #  112)  Sales Comparison by Product (Grocery Shop in Deoria)

 *   You run a small grocery shop in Deoria, and want to compare the sales of products using a bar chart.

*  Topics
   * Vertical Bar Chart

  *  X and Y labels

  *  Title

  *  Color customization

  *  Grid for better readability



```python
import matplotlib.pyplot as plt

Product = ['Rice', 'Wheat', 'Oil', 'Sugar', 'Salt']
sales = [15000, 12000, 8000, 9000, 4000]


plt.bar(Product, sales, color='skyblue')


plt.title("Product Sales in Grocery Shop - Deoria")
plt.xlabel("Product")
plt.ylabel("Sales Rs")
plt.grid(axis='y')

# Show the plot
plt.show()
```

# 113)  Population by Age Group (Pie Chart)

*   Create a pie chart to visualize population distribution across different age group
*    This will help you understand how the population is spread



```python
age_groups = ['0â14', '15â24', '25â44', '45â64', '65+']
population = [25000, 18010, 30020, 20000, 7000]

import matplotlib.pyplot as plt

colors = ['red', 'blue', 'green', 'yellow', 'orange']

#create pie chart
plt.pie(
    population,
    labels=age_groups,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140
)

#add title
plt.title("Population Distribution by Age Group")

#make pie a circle
plt.axis('equal')

#show chart
plt.show()

```

# 114 )Petrol vs Diesel Prices Over a Week

*  You're comparing Petrol and Diesel prices over a week using multiple line plots
*   This helps in visualizing price trends clearly.

*  Topics
   *  Multiple Line Plots

  *  legend ()  
  *  Colors,  Markers, Line Styles

  * Title an d Axis Labels

   *   Grid for better readability

```python
import matplotlib.pyplot as plt


days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
petrol_prices = [97, 98, 99, 100, 100, 101, 102]
diesel_prices = [89, 89.5, 90, 90.5, 91, 91.5, 92]

# create line plots
plt.plot(days, petrol_prices, marker='o', linestyle='-', color='red', label='Petrol')
plt.plot(days, diesel_prices, marker='s', linestyle='--', color='green', label='Diesel')

# Add title and labels
plt.title("Fuel Prices Over a Week")
plt.xlabel("Day")
plt.ylabel("Price (â¹ per litre)")

# Add legend to differentiate lines
plt.legend()

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

#  115)  App Ratings vs Downloads (Scatter Plot)

 *  You have data of mobile apps showing their ratings and download count

 *  Topics
  *   scatter() plot

   *  Custom marker size and color

    *   Title, axis labels

    *   Optional Annotate app names on points



```python
import matplotlib.pyplot as plt

apps = ['Youtube', 'GBU', 'Byjus', 'Github', 'leet_code']
ratings = [4.1, 4.5, 3.8, 4.7, 4.0]
downloads = [500000, 1500, 30000, 200000, 70000]

plt.scatter(ratings, downloads, color='purple', s=100)


plt.title("App Ratings vs Downloads")
plt.xlabel("Rating")
plt.ylabel("Downloads")


for i in range(len(apps)):
    plt.text(ratings[i], downloads[i], apps[i])

# Show the plot
plt.grid(True)
plt.show()
```
