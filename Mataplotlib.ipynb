  # 111) Weather Temperature Trend (Gorakhpur)

*   Plot the temperature trend over a week in Gorakhpur using a line graph

*   This helps visualize how the temperature changes day by day.
*   Topics
     *  Line plot

    *  Labels (xlabel, ylabel)

   *   Title

    *   Grid

    *  Color and Marker style



```python
import matplotlib.pyplot as plt

#Data
days =['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature= [32, 34, 35, 33, 31, 30, 29]


plt.plot(days, temperature, color='red', marker='o' , linestyle='-', linewidth=2)


plt.title("Weekly Temperature Trend - Gorakhpur")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")

plt.grid(True)

plt.show()
```

 #  112)  Sales Comparison by Product (Grocery Shop in Deoria)

*  Topics
   * Vertical Bar Chart

  *  X and Y labels

  *  Title

# 113)  Population by Age Group (Pie Chart)

*   Create a pie chart to visualize population distribution across different age group
*    This will help you understand how the population is spread



```python
age_groups = ['0â14', '15â24', '25â44', '45â64', '65+']
population = [25000, 18010, 30020, 20000, 7000]

import matplotlib.pyplot as plt

colors = ['red', 'blue', 'green', 'yellow', 'orange']

#create pie chart
plt.pie(
    population,
    labels=age_groups,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140
)

#add title
plt.title("Population Distribution by Age Group")

#make pie a circle
plt.axis('equal')

#show chart
plt.show()

```

# 114 )Petrol vs Diesel Prices Over a Week

*  You're comparing Petrol and Diesel prices over a week using multiple line plots
*   This helps in visualizing price trends clearly.

*  Topics
   *  Multiple Line Plots

  *  legend ()  
  *  Colors,  Markers, Line Styles

  * Title an d Axis Labels

   *   Grid for better readability

```python
import matplotlib.pyplot as plt


days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
petrol_prices = [97, 98, 99, 100, 100, 101, 102]
diesel_prices = [89, 89.5, 90, 90.5, 91, 91.5, 92]

# create line plots
plt.plot(days, petrol_prices, marker='o', linestyle='-', color='red', label='Petrol')
plt.plot(days, diesel_prices, marker='s', linestyle='--', color='green', label='Diesel')

# Add title and labels
plt.title("Fuel Prices Over a Week")
plt.xlabel("Day")
plt.ylabel("Price (â¹ per litre)")

# Add legend to differentiate lines
plt.legend()

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

#  115)  App Ratings vs Downloads (Scatter Plot)

 *  You have data of mobile apps showing their ratings and download count

 *  Topics
  *   scatter() plot

   *  Custom marker size and color

    *   Title, axis labels

    *   Optional Annotate app names on points



```python
import matplotlib.pyplot as plt

apps = ['Youtube', 'GBU', 'Byjus', 'Github', 'leet_code']
ratings = [4.1, 4.5, 3.8, 4.7, 4.0]
downloads = [500000, 1500, 30000, 200000, 70000]

plt.scatter(ratings, downloads, color='purple', s=100)


plt.title("App Ratings vs Downloads")
plt.xlabel("Rating")
plt.ylabel("Downloads")


for i in range(len(apps)):
    plt.text(ratings[i], downloads[i], apps[i])

# Show the plot
plt.grid(True)
plt.show()
```

# 116) Monthly Expenses Breakdown (Advanced Pie Chart)

*  Visualize how your monthly salary is spent on rent, groceries, bills, savings, etc
*  Use a pie chart with better formatting to understand your spending habits

*  Topics
   *  Advanced pie chart

   *  explode to highlight important slice

   *  Custom colors

  *  autopct with formatting

  *  Shadow and startangle

```python
import matplotlib.pyplot as plt

#expense
labels = ['Rent', 'Salaries', 'Marketing', 'Utilities', 'Software']
expenses = [18000, 35000, 12000, 4000, 6000]

# Highlight Salaries slice
explode = (0, 0.1, 0, 0, 0)  # 2nd slice salaries is exploded
colors = ['red', 'orange', 'yellow','blue', 'green']

# Plot Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(expenses, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=colors, explode=explode, shadow=True)
plt.title("Startup Monthly Expense Breakdown")
plt.axis('equal')
plt.show()
```

 # 117)   Stackplot â Hiring Report

*  A company tracks hiring in different departments (like Tech, Sales, HR) month-by-month
*   They want to visualize total and stacked hiring growth across months.

* Topic:

   *  stackplot()
   *  xlabel()
   *  ylabel()
   * legend()
   *  title()
   * colors



```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
tech_hires = [2, 3, 4, 4, 5]
sales_hires = [1, 2, 2, 3, 4]
hr_hires = [1, 1, 2, 2, 3]

# Colors for each department
colors = ['red', 'yellow', 'blue']


plt.figure(figsize=(10, 5))
plt.stackplot(months, tech_hires, sales_hires, hr_hires,
              labels=['Tech', 'Sales', 'HR'],
              colors=colors)

plt.title("Hiring Trends by Department (Monthly)")
plt.xlabel("Month")
plt.ylabel("No. of Hires")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# 118) Employee Age Distribution

*  The HR department wants to understand the age distribution of employees in the company
*  A histogram is the best way to see how many employees fall in each age group.

* Topic	Used
  *  hist()
  *  bins
  * xlabel()
  * ylabel()
  * title()
  * grid()
  * color
 * edgecolor

```python
import matplotlib.pyplot as plt


employee_ages = [22, 25, 26, 24, 28, 30, 32, 35, 36, 38, 40, 42, 44, 45, 50, 52, 55, 57, 60]
plt.figure(figsize=(8, 5))
plt.hist(employee_ages, bins=7, color='skyblue', edgecolor='black')

plt.title("Employee Age Distribution")
plt.xlabel("Age Range")
plt.ylabel("Number of Employees")
plt.grid(axis='y')
plt.show()
```

#  119) Horizontal Bar Chart â Tool Usage in Team

 *  A tech team uses various tools like Slack, Zoom, Jira, etc
 *  Management wants to visualize how often each tool is used by team members
 * Used
   * barh()
  *  xlabel()
  *  title()
  *  color
  *  grid(axis='x')

```python
import matplotlib.pyplot as plt

tools = ['Slack', 'Zoom', 'Jira', 'leetcode', 'GitHub']
usage_counts = [35, 28, 30, 22, 40]

# horizontal bar
plt.figure(figsize=(8, 5))
plt.barh(tools, usage_counts, color='lightgreen')
plt.xlabel("Usage Count")
plt.title("Tool Usage in Tech Team")
plt.grid(axis='x')
plt.show()
```
